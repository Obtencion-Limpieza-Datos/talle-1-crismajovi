```python
#importar librerias
import pandas as pd
import sys
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from pandas.tools.plotting import radviz
import ipywidgets as widgets
import pandas as pd
%matplotlib inline
```

```python
#Unificar los siguientes archivos en un solo dataset:
mi_df = pd.read_csv('data/xaa.csv' , encoding="latin-1")
mi_df2 = pd.read_csv('data/xab.csv', encoding="latin-1")
mi_df3 = pd.read_csv('data/xac.csv', encoding="latin-1")
mi_df4 = pd.read_csv('data/xad.csv', encoding="latin-1")
mi_df5 = pd.read_csv('data/xae.csv', encoding="latin-1")
mi_df6 = pd.read_csv('data/xag.csv', encoding="latin-1")
mi_df7 = pd.read_csv('data/xaf.csv', encoding="latin-1")
```

```python
#concatenar
data3 = pd.concat([mi_df, mi_df2, mi_df3, mi_df4, mi_df5, mi_df6, mi_df7])
```

```python
#datos concatenados
data3.head()
```

```python
#Al atributo peso tranformarlo a un tipo de datos float
data3['peso']= data3['peso'].replace(u"Sin información", "0").astype('float')
```

```python
#Obtener un dataset en base a un cantón de nacimiento del siguiente listado
from IPython.display import display

w = widgets.Dropdown(
    options=['Salinas', 'Santa Elena', 'Guayaquil', 'La libertad', 'Mejia', 'Pangua', 'Sigchos'],
    value='Salinas',
    description='Cantones:',
    disabled=False,
)
display(w)
data3[data3['cant_insc']==w.value]
```

```python
#Obtener un dataset que permita obtener todos los nacimientos mayores al día 15 
#y que pertenezcan a una provincia de la región costa 
#data3['dia_insc']
data3['dia_insc']= data3['dia_insc'].replace(u"Sin información", "0")
data3['dia_insc']= data3['dia_insc'].replace(u" ", "0")
data3['dia_insc']= data3['dia_insc'].astype('int64')
from IPython.display import display
w = widgets.Dropdown(
    options=['Costa', 'Sierra', 'Region Insular'],
    value='Costa',
    description='Regiones:',
    disabled=False,
)
display(w)
#data3[(data3['prov_insc	'] ==w.value)]
data3.query('dia_insc > 15')
```

```python
#Obtener un dataset que permita obtener todos los nacimientos en base a la selección del tipo de parto
#Obtener un dataset en base a un cantón de nacimiento del siguiente listado
from IPython.display import display
w = widgets.Dropdown(
    options=['Loja', 'Pichincha', 'Azuay'],
    value='Azuay',
    description='Provincias:',
    disabled=False,
)
w2 = widgets.Dropdown(
    options=['Normal', u'Cesárea', u'Sin información'],
    value=u'Cesárea',
    description='Tipo Parto:',
    disabled=False,
)
display(w,w2)
data3[(data3['prov_insc'] ==w.value) & (data3['tipo_part']==w2.value)]
```

```python
#Obtener un gráfico que me permita vizualizar los nacimientos en base al atributo tipo_instruccion
plt.figure(figsize=(40,20))
plt.subplot2grid((2,3), (0,0))
data3.niv_inst[data3['anio_insc']=='2018'].value_counts().plot(kind='bar')
plt.title('Nacimientos 2018', fontsize = 30)
plt.subplot2grid((2,3), (0,1))
data3.niv_inst[data3['anio_insc']=='2017'].value_counts().plot(kind='bar')
plt.title('Nacimientos 2017', fontsize = 30)
plt.show()
```
